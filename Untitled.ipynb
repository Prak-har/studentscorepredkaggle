```python
import optuna
import lightgbm as lgb
from sklearn.metrics import mean_squared_error
import numpy as np # Import numpy for sqrt

# 1. DEFINE OBJECTIVE
def objective_lgb(trial):
    # Use the 'X_lgb_tr' split we made earlier (strict categories)
    # If not available, just use traincat but cast to category inside
    X_tr, X_val, y_tr, y_val = train_test_split(traincat, y_train, test_size=0.2, random_state=42)

    params = {
        'n_estimators': 5000,
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 20, 150),
        'max_depth': trial.suggest_int('max_depth', 3, 12),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True),

        # GPU SETTINGS
        'device': 'cpu', # Changed to 'cpu' to avoid CUDA compilation error
        'metric': 'rmse',
        'verbosity': -1,
        'random_state': 42,
        'n_jobs': -1
    }

    # Train
    model = lgb.LGBMRegressor(**params)

    # Callback for pruning
    pruning_callback = optuna.integration.LightGBMPruningCallback(trial, "rmse")

    model.fit(
        X_tr, y_tr,
        eval_set=[(X_val, y_val)],
        eval_metric='rmse',
        callbacks=[pruning_callback],
        categorical_feature='auto'
    )

    preds = model.predict(X_val)
    rmse = np.sqrt(mean_squared_error(y_val, preds)) # Changed to use np.sqrt
    return rmse

# 2. RUN STUDY
print("Starting LightGBM Optimization...")
study_lgb = optuna.create_study(direction='minimize')
study_lgb.optimize(objective_lgb, n_trials=100)

print("------------------------------------------------")
print("BEST LIGHTGBM PARAMS:", study_lgb.best_params)
print("BEST RMSE:", study_lgb.best_value)
print("------------------------------------------------")
```
